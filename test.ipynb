```python
class HeapItem:
    def __init__(self, count, pair):
        self.count = count
        self.pair = pair

    def __lt__(self, other):
        """
        Custom comparison for the min-heap.
        __lt__ means "less than". The item that is "less than" another
        will have higher priority and be popped first.
        """
        # If counts are different, the one with the HIGHER count is "less than" (higher priority)
        if self.count != other.count:
            return self.count > other.count

        # If counts are tied, the one with the lexicographically LARGER pair is "less than"
        # In Python, ('s', 't') > ('a', 't'), so this works directly.
        return self.pair < other.pair

    def __repr__(self):
        """A nice representation for printing."""
        return f"HeapItem(count={self.count}, pair={self.pair})"
    
import heapq

# Create items with same count but different lexicographical order
item1 = HeapItem(5, (b' h', b'is'))
item2 = HeapItem(5, (b' S', b'he'))

# Test which one is "less than" (higher priority)
print(f"item1 > item2: {item1 > item2}")  # Should be False
print(f"item1 < item2: {item1 < item2}")  # Should be True

# Test in heap
heap = []
heapq.heappush(heap, item1)
heapq.heappush(heap, item2)
popped = heapq.heappop(heap)
print(f"First popped: {popped}")  # Should be the (b's', b't') item
```

```python
import timeit
```

```python
setup_code = """
from collections import Counter, defaultdict
li = list(range(100))
"""

func_1 = """
c = Counter(li)
"""

func_2 = """
c = defaultdict(int)
for i in range(len(li)):
    c[li[i]] += 1
"""

func_3 = """
c = {}
for i in range(len(li)):
    c[li[i]] = c.get(li[i], 0) + 1
"""

time_counter = timeit.timeit(stmt=func_1, setup=setup_code, number=5000000)
time_defaultdict = timeit.timeit(stmt=func_2, setup=setup_code, number=5000000)
time_dict = timeit.timeit(stmt=func_3, setup=setup_code, number=5000000)

print(f"Counter: {time_counter:.4f} seconds")
print(f"Defaultdict: {time_defaultdict:.4f} seconds")
print(f"Dict: {time_dict:.4f} seconds")
```
